Utilizando biblioteca para raspar noticias

```python
from newspaper import Article

import nltk
nltk.download('punkt')

url = 'https://www.jequienews.com/noticia/jequie-atinge-marca-de-30-dias-sem-crimes-violentos-apos-troca-de-comando-na-pm-e-operacoes-da-pc'
article = Article(url, language='pt') # 'pt' para português, 'de' para alemão
article.download()
article.parse()

print("Título:", article.title)
print("Texto:", article.text[:]) # Mostra os primeiros 200 caracteres
```
